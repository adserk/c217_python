```python
def take_order():
    print("Welcome to the Burger Shop!")
    name = input("What is your name? ")
    print('Hello', name)
    combo1 = ["Chicken Burger", "Fries", "Coke", 'Ketchup', 'Small']
    combo2 = ["Beef Burger", "Fries", 'Coke', 'Ketchup', 'Large']
    combo3 = ["Chicken Burger", "Onion Rings", "Diet Coke",'Bbq', 'Medium' ]
    combo4 = ["Beef Burger", "Garden Salad", 'Diet Coke','Bbq', 'Large' ]
    combo_lst = [combo1, combo2, combo3, combo4]

    user_combo = input ('Would you like a combo, yes or no?')
    if user_combo == 'yes':
        print("Choose from the following Combos:")
        for count, value in enumerate( combo_lst):
            print ( str(count) + ". " + str(value) )
        user_combo_input = input ('Which combo would you like? Enter a number: ')
        chosen_combo = combo_lst[int(user_combo_input)]
        print("Your chosen combo is: ", chosen_combo)
    elif user_combo == 'no':
        chosen_combo =[]
        burger = input('Do you want a Chicken or Beef Burger?')
        if burger == "chicken" or burger == "Chicken Burger":
            chosen_combo.append("Chicken Burger")
        if burger == "beef" or burger == "Beef burger":
            chosen_combo.append("Beef Burger")

        side = input('Do you want Fries, Onion Rings or Garden Salad')
        if side == 'fries' or side == 'Fries':
            chosen_combo.append("Fries")
        elif side == 'onion rings' or side == 'Onion Rings':
            chosen_combo.append('Onion Rings')
        elif side == 'garden salad' or side == 'Garden Salad':
            chosen_combo.append('Garden Salad')

        drink = input('Do you want coke or diet coke?')
        if drink == 'coke' or drink == 'Coke':
            chosen_combo.append('Coke')
        elif drink == 'diet coke':
            chosen_combo.append('Diet Coke')
        print("Your chosen Combo is: ", chosen_combo)

        condiment = input ("Would you like ketchup or bbq?")
        if condiment == "ketchup":
            chosen_combo.append(condiment)
        if condiment == 'Bbq':
            chosen_combo.append(condiment)

        size = input ("Would you like Small, Medium or Large Drink?")
        if size == 'small' or size == 'Small':
            chosen_combo.append('Small')
        elif size == 'medium' or size == 'Medium':
            chosen_combo.append('Medium')
        elif size == 'large' or size == 'Large':
            chosen_combo.append('Large')
        print("Your Chosen Combo is: ", chosen_combo)
    return chosen_combo
```
